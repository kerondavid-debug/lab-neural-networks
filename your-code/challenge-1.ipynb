# Challenge 1 - Tic Tac Toe

In this lab you will perform deep learning analysis on a dataset of playing [Tic Tac Toe](https://en.wikipedia.org/wiki/Tic-tac-toe).

There are 9 grids in Tic Tac Toe that are coded as the following picture shows:

In the first 9 columns of the dataset you can find which marks (`x` or `o`) exist in the grids. If there is no mark in a certain grid, it is labeled as `b`. The last column is `class` which tells you whether Player X (who always moves first in Tic Tac Toe) wins in this configuration. Note that when `class` has the value `False`, it means either Player O wins the game or it ends up as a draw.

Follow the steps suggested below to conduct a neural network analysis using Tensorflow and Keras. You will build a deep learning model to predict whether Player X wins the game or not.

## Step 1: Data Engineering

This dataset is almost in the ready-to-use state so you do not need to worry about missing values and so on. Still, some simple data engineering is needed.

1. Read `tic-tac-toe.csv` into a dataframe.
1. Inspect the dataset. Determine if the dataset is reliable by eyeballing the data.
1. Convert the categorical values to numeric in all columns.
1. Separate the inputs and output.
1. Normalize the input data.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow import keras
from tensorflow.keras import layers
```

```python
df = pd.read_csv('/content/tic-tac-toe.csv')
```

```python
# Inspecting data
df.shape
```

```python
df.dtypes
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

for col in df.columns:
  le = LabelEncoder()
  df[col] = le.fit_transform(df[col])
```

```python
X = df.drop("class", axis=1)
Y = df["class"]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

## Step 2: Build Neural Network

To build the neural network, you can refer to your own codes you wrote while following the [Deep Learning with Python, TensorFlow, and Keras tutorial](https://www.youtube.com/watch?v=wQ8BIBpya2k) in the lesson. It's pretty similar to what you will be doing in this lab.

1. Split the training and test data.
1. Create a `Sequential` model.
1. Add several layers to your model. Make sure you use ReLU as the activation function for the middle layers. Use Softmax for the output layer because each output has a single lable and all the label probabilities add up to 1.
1. Compile the model using `adam` as the optimizer and `sparse_categorical_crossentropy` as the loss function. For metrics, use `accuracy` for now.
1. Fit the training data.
1. Evaluate your neural network model with the test data.
1. Save your model as `tic-tac-toe.model`.

```python
# Splitting
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Model building
model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(9,)),
    layers.Dense(32, activation='relu'),
    layers.Dense(2, activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python

model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)
```

```python
model.save("tic-tac-toe.keras")
```

## Step 3: Make Predictions

Now load your saved model and use it to make predictions on a few random rows in the test dataset. Check if the predictions are correct.

```python


# Select a few random rows from X_test
num_predictions = 5
random_indices = np.random.choice(X_test.shape[0], num_predictions, replace=False)
sample_X_test = X_test[random_indices]
sample_y_test = y_test.iloc[random_indices]

# Make predictions
# model.predict returns probabilities for each class
predictions_proba = model.predict(sample_X_test)
predicted_classes = np.argmax(predictions_proba, axis=1) # Get the class with the highest probability

print(f"\nRandom sample from X_test (scaled):\n{sample_X_test}")
print(f"\nTrue labels for the sample:\n{sample_y_test.values}")
print(f"\nPredicted probabilities for the sample:\n{predictions_proba}")
print(f"\nPredicted classes for the sample:\n{predicted_classes}")

# Compare true labels with predicted classes
comparison = pd.DataFrame({
    'True_Label': sample_y_test.values,
    'Predicted_Class': predicted_classes,
    'Correct_Prediction': (sample_y_test.values == predicted_classes)
})
print("\nComparison of True vs. Predicted Classes:")
display(comparison)
```

## Step 4: Improve Your Model

Did your model achieve low loss (<0.1) and high accuracy (>0.95)? If not, try to improve your model.

But how? There are so many things you can play with in Tensorflow and in the next challenge you'll learn about these things. But in this challenge, let's just do a few things to see if they will help.

* Add more layers to your model. If the data are complex you need more layers. But don't use more layers than you need. If adding more layers does not improve the model performance you don't need additional layers.

* Adjust the number of epochs when you fit the training data to the model. Your model performance continues to improve as you train more epochs. But eventually it will reach the ceiling and the performance will stay the same.

### Step 4: Improving the Model

I will attempt to improve the model by:
1. Adding an additional dense layer.
2. Specifying a custom learning rate for the Adam optimizer.
3. Increasing the number of epochs for training.

```python
print("\n--- Step 4: Improve Your Model ---")

# Model building with additional layers
model_improved = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(9,)), # Increased neurons
    layers.Dense(64, activation='relu'),
    layers.Dense(32, activation='relu'), # Added another layer
    layers.Dense(2, activation='softmax')
])
```

```python
# Adjusting the learning rate
custom_adam = keras.optimizers.Adam(learning_rate=0.001) # Example learning rate

model_improved.compile(optimizer=custom_adam, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Fitting the improved model with more epochs
print("\nTraining the improved model...")
model_improved.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2)
```

```python
# Evaluate the improved model
loss, accuracy = model_improved.evaluate(X_test, y_test)
print(f"\nImproved Model Test Loss: {loss:.4f}")
print(f"Improved Model Test Accuracy: {accuracy:.4f}")
```

Now, let's make predictions with the improved model to see if the predictions are more accurate.

```python
# Make predictions with the improved model

# Select a few random rows from X_test
num_predictions = 5
random_indices = np.random.choice(X_test.shape[0], num_predictions, replace=False)
sample_X_test = X_test[random_indices]
sample_y_test = y_test.iloc[random_indices]

predictions_proba_improved = model_improved.predict(sample_X_test)
predicted_classes_improved = np.argmax(predictions_proba_improved, axis=1)

print(f"\nRandom sample from X_test (scaled):\n{sample_X_test}")
print(f"\nTrue labels for the sample:\n{sample_y_test.values}")
print(f"\nPredicted probabilities (Improved Model):\n{predictions_proba_improved}")
print(f"\nPredicted classes (Improved Model):\n{predicted_classes_improved}")

comparison_improved = pd.DataFrame({
    'True_Label': sample_y_test.values,
    'Predicted_Class_Improved': predicted_classes_improved,
    'Correct_Prediction': (sample_y_test.values == predicted_classes_improved)
})
print("\nComparison of True vs. Predicted Classes (Improved Model):")
display(comparison_improved)
```
